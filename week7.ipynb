```python
# confidence interval for populaÔøΩon mean ‚Äì When the populaÔøΩon standard deviaÔøΩon (ùùàùùà) is known. 
# Use the data below to calculate a confidence interval for the true populaÔøΩon mean ÔøΩme (in minutes) 
# students take to commute to College, using a sample of 39 students. Assume the populaÔøΩon 
# standard deviaÔøΩon is known to be 5.6 minutes. 
# 12, 12, 13, 13, 15, 16, 17, 22, 23, 25, 26, 27, 28, 28, 29, 17, 22, 23, 25, 26, 27, 28, 28, 29, 13, 15, 16, 
# 17, 22, 23, 25, 26, 27, 28, 28, 29, 17, 22, 23 
import numpy as np
from scipy.stats import norm


#sample data
x=np.array([12, 12, 13, 13, 15, 16, 17, 22, 23, 25, 26, 27, 28, 28, 29, 17, 22, 23, 25, 26, 27, 28, 28, 29, 13, 15, 16, 
17, 22, 23, 25, 26, 27, 28, 28, 29, 17, 22, 23 ])

#population standar deviation
populationStdDev=5.6

#point estimate(mean)
sampleSize=len(x)
pointEstimate=np.mean(x)    

#critical value based on z distribution for 95% confidence interval
confidenceLevel=0.95
zCritical=norm.ppf((1+confidenceLevel)/2)

#standar Error
standarError=populationStdDev/np.sqrt(len(x))

#margin of error
marginError=zCritical*standarError

#confidence interval
lowerBound=pointEstimate-marginError
upperBound=pointEstimate+marginError

#output results
print(f'Sample size:{sampleSize}')
print(f'Point estimate:{pointEstimate}')
print(f'Critical value (z):{zCritical}')
print(f'Standar Error:{standarError}')
print(f'95% Confidence Interval:[{lowerBound}, {upperBound}]')



```

```python
# Confidence Intervals Using the Students t DistribuÔøΩon. 
# If we‚Äôre working with smaller samples (n<30), the ÔøΩmes follow a normal distribuÔøΩon and populaÔøΩon 
# variance is not known, we use the Student‚Äôs t DistribuÔøΩon. 
# Use the data below to calculate a confidence interval for the true populaÔøΩon mean ÔøΩme (in minutes) 
# students take to commute to College, using a sample of 39 students. Assume the populaÔøΩon 
# standard deviaÔøΩon is Unknown and the data comes from a normally distributed populaÔøΩon.
# 12, 12, 13, 13, 15, 16, 17, 22, 23, 25, 26, 27, 28, 28, 29
import numpy as np
from scipy.stats import t 

#sample data
x=np.array([12, 12, 13, 13, 15, 16, 17, 22, 23, 25, 26, 27, 28, 28, 29])

#sample size
sampleSize=len(x)

#degrees of freedom for t-distribution
df=sampleSize-1

#point estimate(mean)
pointEstimate=np.mean(x)

#critical value based on t-distribution for 95% confidence interval
confidenceLevel=0.95
tCritical=t.ppf((1+confidenceLevel)/2,df)

#standar error
std_dev=np.std(x,ddof=1) #ddof+1  for sample standard deviation
standarError=std_dev/np.sqrt(sampleSize)

#margin error
marginError=tCritical*standarError

#confidence interval

lowerBound=pointEstimate-marginError
upperBound=pointEstimate+marginError


# output results





```
